## Element wise Operations

```python
x = torch.tensor([[1,2,3],[4,5,6]])
y = torch.tensor([[n for n in range(4,7)],[n for n in range(1,4)]])

#ADD
print(x+y)

#SUB
print(x-y)

#MUL
print(x*y)

#DIV
print(x/y)

#POWER
print(x**y)

#abs
print(abs(x-y))

#negative
print(torch.neg(abs(x-y)))

#round, floor, ceil

#clamp
torch.clamp(x,min=1,max=3)
```

## Reduction Operation

```python
#Sum
a = torch.randint(size=(2,3),low=0, high=7, dtype=torch.float32)
print(a)
print(torch.sum(a))

#Sum along columns
print(torch.sum(a,dim=0))

#Sum along Rows
print(torch.sum(a,dim=1))
```

```python
#Mean
print("Mean--> ",torch.mean(a))

#Median
print("Median--> ", torch.median(a))
```

```python
#Max and min
print("Max--> ", torch.max(a))
print("Min--> ", torch.min(a))
```

```python
#Product
print("Product--> ", torch.prod(a))

#Standard Deviation
print("Standard deviation--> ", torch.std(a))

#Variance
print("Variance--> ", torch.var(a))
```

```python
#arg max and arg min
print("Max Item at--> ",torch.argmax(a))
print("Min Item at--> ", torch.argmin(a))
```
