## Element wise Operations

```python
x = torch.tensor([[1,2,3],[4,5,6]])
y = torch.tensor([[n for n in range(4,7)],[n for n in range(1,4)]])

#ADD
print(x+y)

#SUB
print(x-y)

#MUL
print(x*y)

#DIV
print(x/y)

#POWER
print(x**y)

#abs
print(abs(x-y))

#negative
print(torch.neg(abs(x-y)))

#round, floor, ceil

#clamp
torch.clamp(x,min=1,max=3)
```

## Matrix Operations

```python
f = torch.randint(size=(3,3), low = 1, high = 10, dtype=torch.float32)
g = torch.randint(size=(3,3), low = 1, high = 10, dtype=torch.float32)

print(f)
print(g)
```

```python
#Matrix Multiplication
torch.matmul(f,g)
```

```python
#Dot product
vector1 = torch.randint(size=(5,),low=1, high=10)
vector2 = torch.randint(size=(5,),low=1, high=10)
torch.dot(vector1, vector2)
```

```python
#Transpose
torch.transpose(f, 0, 1)
```

```python
#Determinant
torch.det(f)
```

```python
torch.inverse(f)
```

## Tensor Operations on a GPU

```python
torch.cuda.is_available()
```

```python
device = torch.device('cuda')
```

```python
#Making a new tensor on our GPU
torch.rand((2,3), device=device)
```

```python
#Moving to a GPU
a = torch.rand(2,3)
print(a)
b = a.to(device)
print(b)
```

```python
# Comparision between GPU and CPU
import time

a = torch.rand(10000,10000)
b = torch.rand(10000,10000)

start_time_cpu = time.time()
result_cpu = torch.matmul(a,b)
end_time_cpu = time.time()
print("CPU Time--> ", end_time_cpu-start_time_cpu)

c = a.to(device)
d = b.to(device)

start_time_gpu = time.time()
result_gpu = torch.matmul(c,d)
end_time_gpu = time.time()
print("CPU Time--> ", end_time_gpu-start_time_gpu)
```
